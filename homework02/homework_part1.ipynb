### Task II: the game of life

Now it's time for you to make something more challenging. We'll implement Conway's [Game of Life](http://web.stanford.edu/~cdebs/GameOfLife/) in _pure pytorch_. 

While this is still a toy task, implementing game of life this way has one cool benefit: __you'll be able to run it on GPU! __ Indeed, what could be a better use of your gpu than simulating game of life on 1M/1M grids?

![img](https://cdn.tutsplus.com/gamedev/authors/legacy/Stephane%20Beniak/2012/09/11/Preview_Image.png)
If you've skipped the url above out of sloth, here's the game of life:
* You have a 2D grid of cells, where each cell is "alive"(1) or "dead"(0)
* Any living cell that has 2 or 3 neighbors survives, else it dies [0,1 or 4+ neighbors]
* Any cell with exactly 3 neighbors becomes alive (if it was dead)

For this task, you are given a reference numpy implementation that you must convert to pytorch.
_[numpy code inspired by: https://github.com/rougier/numpy-100]_


__Note:__ You can find convolution in `torch.nn.functional.conv2d(Z,filters)`. Note that it has a different input format. 


````python
from scipy.signal import convolve2d

def np_update(Z):
    # Count neighbours with convolution
    filters = np.array([[1,1,1],
                        [1,0,1],
                        [1,1,1]])
    
    N = convolve2d(Z,filters,mode='same')
    
    # Apply rules
    birth = (N==3) & (Z==0)
    survive = ((N==2) | (N==3)) & (Z==1)
    
    Z[:] = birth | survive
    return Z
````

````python
def torch_update(Z):
    """
    Implement an update function that does to Z exactly the same as np_update.
    :param Z: torch.FloatTensor of shape [height,width] containing 0s(dead) an 1s(alive)
    :returns: torch.FloatTensor Z after updates.
    
    You can opt to create new tensor or change Z inplace.
    """
    
    weights = torch.from_numpy(
        np.array([
            [1, 1, 1],
            [1, 0, 1],
            [1, 1, 1]
        ])
    ).reshape(
        (1, 1, 3, 3)
    ).type(
        torch.FloatTensor
    )
    
    N = torch.nn.functional.conv2d(
        Z.reshape((1, 1, Z.shape[0], Z.shape[1])),
        weights,
        padding=1
    )
    
    birth = (N == 3) & (Z == 0)
    survive = ((N == 2) | (N == 3)) & (Z == 1)
    
    Z[:] = (birth | survive).data
    
    return Z
````

````python
#initial frame
Z_numpy = np.random.choice([0,1],p=(0.5,0.5),size=(100,100))
Z = torch.from_numpy(Z_numpy).type(torch.FloatTensor)

#your debug polygon :)
Z_new = torch_update(Z.clone())

#tests
Z_reference = np_update(Z_numpy.copy())
assert np.all(Z_new.numpy() == Z_reference), "your pytorch implementation doesn't match np_update. Look into Z and np_update(ZZ) to investigate."
print("Well done!")
````

### Let's go

````python
from torch.autograd import Variable
from IPython.display import clear_output
````

````python
def get_batches(X, y, batch_size):
    size = len(X)
    
    indices = np.arange(size)
    np.random.shuffle(indices)
    
    epoch_accuracy = 0
    epoch_losses = []
    
    for start in range(0, size, batch_size):
        end = min(start + batch_size, size)
        batch_idx = indices[start:end]
        
        x_batch = Variable(torch.FloatTensor(X[batch_idx]))
        y_batch = y[batch_idx]
        
        yield x_batch, y_batch
````

````python
n_samples = len(X_train)

batch_size = 8
learning_rate = 0.1
n_epochs = 50

momentum = 0.9

init_scale = 1e-3

hidden_size = 50 # size of hidden layer
````

````python
# first dense layer
w1 = Variable(torch.randn(784, hidden_size) * init_scale, requires_grad=True)
b1 = Variable(torch.randn(hidden_size) * init_scale, requires_grad=True)

# there is sigmoid layer between them

# second dense layer
w2 = Variable(torch.randn(hidden_size, 10) * init_scale, requires_grad=True)
b2 = Variable(torch.randn(10) * init_scale, requires_grad=True)

for epoch in range(n_epochs):
    
    epoch_accuracy = 0
    epoch_losses = []
    
    for x_batch, y_batch in get_batches(X_train, y_train, batch_size):

        # simple logistic
        #output = torch.nn.functional.softmax(torch.matmul(x_batch, w1) + b1)
        
        output = torch.nn.functional.log_softmax(
            torch.matmul(1 / (1 + torch.exp(-( torch.matmul(x_batch, w1) + b1))), w2) + b2,
            dim=1
        )
        
        y_predicted = np.argmax(output.data.numpy(), axis=1)
        epoch_accuracy += np.sum(y_predicted == y_batch)
         
        # calc loss
        loss = torch.nn.functional.nll_loss((output), torch.from_numpy(y_batch))
        loss.backward()
        
        epoch_losses.append(float(loss))
        
        # update weights with backprop
        w1.data -= learning_rate * w1.grad.data
        b1.data -= learning_rate * b1.grad.data
        
        w2.data -= learning_rate * w2.grad.data
        b2.data -= learning_rate * b2.grad.data
        
        # momentum
        w1.grad.data *= momentum
        b1.grad.data *= momentum
        
        w2.grad.data *= momentum
        b2.grad.data *= momentum
        
    # for logistict
    #test_output = torch.nn.functional.softmax(torch.matmul(torch.FloatTensor(X_test), w1) + b1)
       
    test_output = torch.matmul(
        1 / (1 + torch.exp(-(torch.matmul(torch.FloatTensor(X_test), w1) + b1))),
        w2
    ) + b2
    
    y_predicted = np.argmax(test_output.data.numpy(), axis=1)
    
    clear_output()
    
    print("Train (epoch %d) accuracy: %f" % (epoch, epoch_accuracy / n_samples))
    print("Epoch mean loss:", np.mean(epoch_losses))
    print("Test accuracy:", np.sum(y_predicted == y_test) / len(y_test))
````
